### Adding both x and y data

You may be wondering why the x-axis ranges from 0-3 and the y-axis from 1-4. If you provide a single list or array to the [plot()](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) command,*** matplotlib assumes it is a sequence of y values, and automatically generates the x values for you.***

[plot()](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot) is a versatile command, and will take an arbitrary number of arguments. For example, to plot x versus y, you can issue the command:

```python
x_vals = [1,2,3,4]
y_vals = [1, 4, 9, 16]
plt.plot(x_vals, y_vals)
plt.show()
```

### Setting axis ranges
You can set the [xmin, xmax, ymin, ymax] of the plot using the [axis()](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.axis) command. So, for example:

```python
import matplotlib.pyplot as plt
plt.plot(x_vals, y_vals)
plt.axis([0, 6, 0, 20])
plt.show()
```

### Resizing the plot
We can also resize the plot using figsize. This will be important later when we make multiple plots together.

```python
import random


num_rand_data_points = 10

#change the figsize values below to see how it change the size of the plot
#Change the facecolor to 'y' to see what part of plot changes Look up the docs to see all possible colors
plt.figure(figsize=(10, 5), facecolor='b')

#Generates random integers between 1 and 49 (inclusive) with a step of 1.
#Notice the use of list comprehension we learn in Python Refresher class
x_vals = [random.randrange(1, 50, 1) for i in range(num_rand_data_points)]

plt.plot(x_vals, label="Random data")
plt.ylabel('The y-axis label!')
plt.xlabel('The x-axis label!')
plt.title("{} random data points!".format(num_rand_data_points))
plt.legend()
plt.show()
```

### Plotting multiple values
You can plot multiple lines too:

```python
# setup the data
(x1_vals, y1_vals) = ([1,2,3,4], [1, 4, 9, 16])
(x2_vals, y2_vals) = ([0,1,1.5,7], [1,4,12,16])
(x3_vals, y3_vals) = ([0,1,3,5], [3,4.5,7,18])

# plot the data
plt.figure(figsize=(10, 5))
plt.plot(x1_vals, y1_vals)
plt.plot(x2_vals, y2_vals)
plt.plot(x3_vals, y3_vals)
plt.show()
```

An even more compact way of doing this:

```python
plt.figure(figsize=(10, 5))
plt.plot(x1_vals, y1_vals, x2_vals, y2_vals, x3_vals, y3_vals)
plt.show()
```

#Putting it into practice :

Let's plot Animal movement in a Zoo!

```python
num_points = 50

#Generate a random element where each number is a random float between 0 and 10 for 50 times ( or num of points times)
x_panda = [random.uniform(0, 10) for _ in range(num_points)]
y_panda = [random.uniform(0, 10) for _ in range(num_points)]

x_dog = [random.uniform(0, 12) for _ in range(num_points)]
y_dog = [random.uniform(0, 12) for _ in range(num_points)]

#Cause there is no such thing as too much dogs
x_dog2 = [random.uniform(0, 8) for _ in range(num_points)]
y_dog2 = [random.uniform(0, 8) for _ in range(num_points)]

# Plotting the animal paths
plt.figure(figsize=(8, 6))

plt.plot(x_panda, y_panda, label='Panda')
plt.plot(x_dog, y_dog, label='Dog')
plt.plot(x_dog2, y_dog2, label='dog2')

plt.title('Animal Movement in the Wild')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.tight_layout()

# plt.show()
```

## Controlling Line Properties
- https://matplotlib.org/users/pyplot_tutorial.html#controlling-line-properties

Lines have many attributes that you can set: linewidth, dash style, marker, antialiasing, alpha, and more; see [matplotlib.lines.Line2D](https://matplotlib.org/api/lines_api.html#matplotlib.lines.Line2D). There are several ways to set line properties

```python
(x1_vals, y1_vals) = ([1,2,3,4], random.sample(range(0, 5), 4))
(x2_vals, y2_vals) = ([1,2,3,4], random.sample(range(0, 5), 4))
(x3_vals, y3_vals) = ([1,2,3,4], random.sample(range(0, 5), 4))

line1 = plt.plot(x1_vals, y1_vals)
line2 = plt.plot(x2_vals, y2_vals)
line3 = plt.plot(x3_vals, y3_vals)
# marker='D': Sets the marker style to a diamond ('D')
# alpha=0.5: Sets the transparency of the line to 50% (0.5 alpha value), making it partially transparent.
plt.setp(line1, color='r', linewidth=9.0, marker='D', alpha=0.5)
plt.setp(line2, color='b', linewidth=2.0, marker='X', alpha=0.5)
plt.show()
```

## Plotting multiple charts

You can create multiple plots within the same figure by using [`subplot`](https://matplotlib.org/gallery/subplots_axes_and_figures/subplot.html)

**References:**
- https://jakevdp.github.io/PythonDataScienceHandbook/04.08-multiple-subplots.html
- https://matplotlib.org/gallery/subplots_axes_and_figures/subplot.html
- https://matplotlib.org/devdocs/gallery/subplots_axes_and_figures/subplots_demo.html

```python
import matplotlib.pyplot as plt

# Sales data for three products across four quarters
products = ['Product A', 'Product B', 'Product C']
quarters = ['Q1', 'Q2', 'Q3', 'Q4']

sales_data = [
    [50, 60, 45, 55],  # Sales for Product A
    [70, 80, 75, 85],  # Sales for Product B
    [30, 40, 35, 45]   # Sales for Product C
]


# 3, 1: Indicates that you want to create 3 subplots vertically (3 rows and 1 column).
# figsize=(8, 10) Specifies the dimensions of the entire figure in inches, where 8 inches is the width, and 10 inches is the height.
fig, axs = plt.subplots(3, 1, figsize=(8, 10))

# Plotting sales data for each product in separate subplots
for i, product in enumerate(products):
    axs[i].plot(quarters, sales_data[i], marker='d', linestyle='-', label=product)
    axs[i].set_title(f'Sales Trend for {product}')
    axs[i].set_xlabel('Quarters')
    axs[i].set_ylabel('Sales')
    axs[i].legend()

# Adjusting layout
plt.tight_layout()
plt.show()
```

- https://matplotlib.org/users/colors.html

In almost all places in matplotlib where a color can be specified by the user it can be provided as:

- an RGB or RGBA tuple of float values in [0, 1] (e.g., (0.1, 0.2, 0.5) or (0.1, 0.2, 0.5, 0.3))
- a hex RGB or RGBA string (e.g., '#0F0F0F' or '#0F0F0F0F')
- a string representation of a float value in [0, 1] inclusive for gray level (e.g., '0.5')
one of {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}
- a X11/CSS4 color name
- a name from the xkcd color survey prefixed with 'xkcd:' (e.g., 'xkcd:sky blue')

##Histograms

```python
import matplotlib.pyplot as plt
import random

# Sample data: Favorite singers of 100 people
singers = ['Taylor Swift', 'Post Malone', 'The Weeknd', 'The Chainsmokers']
votes = [random.choice(singers) for _ in range(100)]

# Count the votes for each singer
counts = [votes.count(singer) for singer in singers]

# Creating the histogram
plt.figure(figsize=(10, 6))
plt.bar(singers, counts, color=['red', 'blue', 'purple', 'green'])
plt.title('Favorite Singer Survey Results')
plt.xlabel('Singer')
plt.ylabel('Number of Votes')
plt.yticks(range(0, max(counts) + 1, 5))
plt.show()
```

##ScatterPlot


```python
import matplotlib.pyplot as plt
import random

# Generate random data for superhero action figures
num_figures = 20

# Random superhero names for labeling points
superhero_names = [
    "Superman", "Batman", "Wonder Woman", "Spider-Man", "Iron Man",
    "Captain America", "Thor", "Black Widow", "Hulk", "Aquaman",
    "Flash", "Green Lantern", "Black Panther", "Wolverine", "Deadpool",
    "Doctor Strange", "Ant-Man", "Captain Marvel", "Hawkeye", "Star-Lord"
]

# Generate random heights and weights for superhero action figures
heights = [random.uniform(150, 200) for _ in range(num_figures)]  # Heights in centimeters
weights = [random.uniform(60, 100) for _ in range(num_figures)]   # Weights in kilograms

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(heights, weights, marker='o', color='blue', s=100)

# Label each point with a random superhero name
for i, name in enumerate(superhero_names):
    plt.annotate(name, (heights[i], weights[i]), fontsize=10, ha='right')

# Add labels and title
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Superhero Action Figures: Height vs. Weight")

# # Highlight the "strongest" superhero
strongest_index = weights.index(max(weights))
strongest_superhero = superhero_names[strongest_index]
plt.scatter(heights[strongest_index], weights[strongest_index], color='red', s=150, label=strongest_superhero)
plt.legend()

plt.grid(True)
plt.tight_layout()

plt.show()
```
